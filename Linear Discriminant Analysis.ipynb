# Linear Discriminant Analysis (LDA)

While PCA attempts to find the orthogonal component axes of maximum variance in a dataset; the goal in LDA is to find the feature subspace that optimizes class separability. In this sense, LDA is a supervised algorithm, and although it can be used as a standalone linear classifier, more commonly it is applied for dimensionality reduction before classification. LDA is used with categorical dependent variables and continuous independent variables, which is the reverse of a similar process, analysis of variance (ANOVA). A key assumption of the LDA method is that the independent variables are normally distributed. 

<img src="extras/05_06.png" width="300" height="300" />

The image above illustrates the concept of LDA: given the two classes (x's and o's), the linear discriminant on the x-axis (LD1) would do a better job of separating the two classes than LD2, which captures a lot of the variance in the dataset but none of the class-discriminatory information.

LDA is not necessarily superior to PCA for multi-class classification. In image recognition, PCA has shown to outperform LDA if the number of samples per class is relatively small. In practice, it is also not uncommon to use both LDA and PCA in combination: E.g., PCA for dimensionality reduction followed by an LDA.

Although we will not go through the exercise of demonstrating the inner workings of the LDA algorithm, these are the main steps that comprise the method:

1 - Compute the dd-dimensional mean vectors for the different classes from the dataset.  
2 - Compute the scatter matrices (in-between-class and within-class scatter matrix).  
3 - Compute the eigenvectors (e1,e2,...,ed) and corresponding eigenvalues (λ1,λ2,...,dλ) for the scatter matrices.  
4 - Sort the eigenvectors by decreasing eigenvalues and choose kk eigenvectors with the largest eigenvalues to form a d×k dimensional matrix W (where every column represents an eigenvector).  
5 - Use this d×k eigenvector matrix to transform the samples onto the new subspace. This can be summarized by the matrix multiplication: Y=X×W (where X is a n×d-dimensional matrix representing the nn samples, and y are the transformed n×k-dimensional samples in the new subspace).  

## Coding Example Using Scikit-Learn
Import the standard libraries

```python
import numpy as np
import pandas as pd
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

We will use the the classic Iris dataset for our example.

```python
iris = datasets.load_iris()
y = iris['target']
X = pd.DataFrame(iris['data'], columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
```

Split the data into training and testing sets.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .33, random_state = 0)
```

First, we need to scale our data.  
_Note_: scaling has been shown to not have an actual impact on LDA results. Although the scale of the eigenvalues will be different, their magnitudes relative to each other will be unaffected. 

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

Let us first take a look at how the variance explained ratio changes as we increase the number of linear discriminants.

```python
lda = LDA(n_components=None)
X_train_lda = lda.fit_transform(X_train_std, y_train)

lda.explained_variance_ratio_
```

As we can see, the first discrimant comprises virtually all of the explained variance ratio.

```python
plt.bar(range(1, 4), lda.explained_variance_ratio_, alpha=0.5, align='center')
plt.step(range(1, 4), np.cumsum(lda.explained_variance_ratio_), where='mid')
plt.ylabel('Explained variance ratio')
plt.xlabel('Linear Discriminants')
plt.show()
```

Now we implement LDA from scikit-learn, choosing 2 as our number of component.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

lda = LDA(n_components=2)
X_train_lda = lda.fit_transform(X_train_std, y_train)
X_test_lda = lda.transform(X_test_std)
```

Looking at our two discriminants along the x- and y-axes, we can see that the two discriminants allow us to separate our three classes somewhat nicely. 

```python
colors = ['r', 'b', 'g']
markers = ['s', 'x', 'o']

for l, c, m in zip(np.unique(y_train), colors, markers):
    plt.scatter(X_train_lda[y_train == l, 0] * (-1),
                X_train_lda[y_train == l, 1] * (-1),
                c=c, label=l, marker=m)

plt.xlabel('LD 1')
plt.ylabel('LD 2')
plt.legend(loc='lower right')
plt.tight_layout()
# plt.savefig('./figures/lda2.png', dpi=300)
plt.show()
```

Now, let's implement a logit model on our linear discriminants to see how well we can predict the target labels.

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr = lr.fit(X_train_lda, y_train)
yhat = lr.predict(X_train_lda)
```

Solely looking at the training data, we see a precision of 94%, which is pretty good. 

```python
from sklearn.metrics import confusion_matrix, classification_report

print confusion_matrix(y_train, yhat)
print '\n', classification_report(y_train, yhat)
```

Now, looking at the test data, we can see that our accuracy goes down to 87%, meaning that our model slighly overfitted to the linear discriminants. 

```python
yhat = lr.predict(X_test_lda)

print confusion_matrix(y_test, yhat)
print '\n', classification_report(y_test, yhat)
```
