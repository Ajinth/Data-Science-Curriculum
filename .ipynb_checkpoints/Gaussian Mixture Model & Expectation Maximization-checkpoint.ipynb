### The Gaussian Mixture Model

The Gaussian Mixture Model (GMM) is a clustering algorithm that allows us to deal with not only overlapping clusters, but also with clusters that have different shapes and density. 

One nice benefit of the GMM is that its output is a probability model, meaning we can use it to sample data, compare two collections of data (like the training and test data), or impute missing values. 

We begin with several mixture components, indexed by *k* (for cluster). Each has a mean $\mu_k$, a variance $\Sigma_k$, and size $\pi_k$. 

\begin{align}
\large \mu_k = \frac{1}{N_k}\displaystyle\sum_{n=1}^N \gamma_k^{(n)}\vec{x}^{(n)}
\end{align}

\begin{align}
\large \Sigma_k = \frac{1}{N_k}\displaystyle\sum_{n=1}^{N}\gamma_k^{(n)}(\vec{x}^{(n)}-\mu_k)(\vec{x}^{(n)}-\mu_k)^T
\end{align}

\begin{align}
\large \pi_k = \frac{N_k}{N} \text{ with }N_k = \displaystyle\sum_{n=1}^N \gamma_k^{(n)}
\end{align}

Our data is described by a latent variable *Z*, which has a $\pi_k$ probability of belonging to a cluster *k*, which has its own probability distribution. The sum of all the individual $\pi$ probabilities is equal to one. 

This is known as a latent variable model. The presence of the latent variable *Z* helps explain the patterns in X, in this case the groups or clusters. Each observation x is representative of an observation *z* of the latent variable, and it has a $\large\gamma_k^{(n)}$ probability of belonging to cluster *k* (the posterior). 

\begin{align}
\renewcommand{\vec}[1]{\mathbf{#1}}
\large \gamma_k^{(n)} = p(z^{(n)}\mid \vec{x}) = \frac{\pi_k \mathcal{N}(\vec{x}^{(n)}\mid\mu_k,\Sigma_k)}{\sum_{j=1}^K\pi_j \mathcal{N}(\vec{x}^{(n)}\mid\mu_j,\Sigma_j)}
\end{align}

### Expectation Maximization

In order to apply the GMM on our data, we will use the Expectation Maximization (EM) algorithm to obtain our cluster membership probabilitites and individual distribution parameters. Expectation Maximization is an iterative method to find maximum likelihood or maximum a posteriori (MAP) estimates of parameters in statistical models, where the model depends on unobserved latent variables. 

In the first step, EM first treats the mean, covariance, and cluster probabilities as fixed. Given these parameters, we compute the probability that it belongs to cluster k. Our goal is to maximize the log-likelihood of our objective function, which is:

\begin{align}
\large \text{log } p(X) = \sum_i log \left[\displaystyle\sum_{j=1}^K\pi_j\mathcal{N}(\vec{x}^{(n)}\mid\mu_k,\Sigma_k)\right]
\end{align}

In the second step, we fix the probabilities r, and then update the parameters that maximize the expected log-likelihood found in the prior step. These parameter-estimates are then used to determine the distribution of the latent variables in the next E step. 

Convergence is not guaranteed to arrive at a global optimimum because the cost function is non-convex and thus susceptible to local optima, so we may have to start at several initializations, and use the log likelihood to arrive at the best. 

The output for each observation is soft cluster membership, which can then be used for hard membership classification.

### GMM Coding Example

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
def gmm(X, K, max_iter=20, smoothing=10e-3):
    N, D = X.shape
    M = np.zeros((K, D))
    R = np.zeros((N, K))
    C = np.zeros((K, D, D))
    pi = np.ones(K) / K # uniform

    # initialize M to random, initialize C to spherical with variance 1
    for k in xrange(K):
        M[k] = X[np.random.choice(N)]
        C[k] = np.eye(D)

    costs = np.zeros(max_iter)
    weighted_pdfs = np.zeros((N, K)) # we'll use these to store the PDF value of sample n and Gaussian k
    for i in xrange(max_iter):
        # step 1: determine assignments / resposibilities
        for k in xrange(K):
            for n in xrange(N):
                weighted_pdfs[n,k] = pi[k]*multivariate_normal.pdf(X[n], M[k], C[k])

        for k in xrange(K):
            for n in xrange(N):
                R[n,k] = weighted_pdfs[n,k] / weighted_pdfs[n,:].sum()

        # a faster way to do step 1: "vectorization"
        # for k in xrange(K):
        #     weighted_pdfs[:,k] = pi[k]*multivariate_normal.pdf(X, M[k], C[k])
        # R = weighted_pdfs / weighted_pdfs.sum(axis=1, keepdims=True)

        # step 2: recalculate params
        for k in xrange(K):
            # same as k-means
            Nk = R[:,k].sum()
            pi[k] = Nk / N
            M[k] = R[:,k].dot(X) / Nk
            # we are adding regularization in order to avoid singular covariance
            C[k] = np.sum(R[n,k]*np.outer(X[n] - M[k], X[n] - M[k]) for n in xrange(N)) / Nk + np.eye(D)*smoothing


        costs[i] = np.log(weighted_pdfs.sum(axis=1)).sum()
        if i > 0:
            if np.abs(costs[i] - costs[i-1]) < 0.1:
                break
    
    plt.figure(figsize=(8,6))
    plt.plot(costs)
    plt.title("Costs")
    plt.show()

    random_colors = np.random.random((K, 1))
    colors = R.dot(random_colors)
    plt.figure(figsize=(8,6))
    plt.scatter(X[:,0], X[:,1], c=colors)
    plt.title('After GMM Clustering')
    plt.show()

    print "pi:", pi
    print "means:", M
    print "covariances:", C
    return
```

```python
# assume 3 means
D = 2 # so we can visualize it more easily
s = 4 # separation so we can control how far apart the means are
mu1 = np.array([0, 0])
mu2 = np.array([s, s])
mu3 = np.array([0, s])

N = 2000 # number of samples
X = np.zeros((N, D))
X[:1200, :] = np.random.randn(1200, D)*2 + mu1
X[1200:1800, :] = np.random.randn(600, D) + mu2
X[1800:, :] = np.random.randn(200, D)*0.5 + mu3

# what does it look like without clustering?
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1])
plt.title('Unclustered Data')
plt.show()

K = 3
gmm(X, K)
```

### Soft (Fuzzy) K-Means

Soft k-means can be thougth of a Gaussian Mixture Model where the cluster probabilities are all identical. In fuzzy k-means, cluster membership probabilities (also known as responsibilities) for each observation are determined by a distance metric between the observation and the mean of the cluster. The $\beta$ parameter allows us to determin the "softness" of the cluster membership; that is, as $\beta$ approaches infinity, our model approaches a hard k-means algorithm.

\begin{align}
\large r_k^{(n)} = \frac{exp[-\beta d(\vec{m}_k,\vec{x}^{(n)})]}{\sum_j exp[-\beta d(\vec{m}_j,\vec{x}^{(n)})]}
\end{align}

The mean of each cluster is calculated as the average probability vector for the given cluster.

\begin{align}
\large \vec{m}_k = \frac{\sum_n r_k^{(n)}\vec{x}^{(n)}}{\sum_n r_k^{(n)}}
\end{align}

The main steps in a Soft K-Means algorithm can be summarized as follows:
1. Initialize $\vec{m}_K$ random points in $\vec{X}$
2. While not converged:
    3. Step 1: Calculate cluster responsibilities.
    4. Step 2: Recalculate cluster means.  
    
Our objective function is merely a distance metric (Euclidian here) that's weighted by responsibility of each observation with respect to a given cluster. The approach here is known as "coordinate descent," meaning we only update one variable at a time. Although there is a mathematical guarantee that J will always decrease with each iteration, it might not find a global minimum. Here is our objective function:

\begin{align}
\ J = \sum_n\sum_kr_k^{(n)}\mid\mid\vec{m}_k-\vec{x}^{(n)}\mid\mid^2
\end{align}

### Soft K-Means Coding Example

First, let's create a function to generate our data. Our data will be 2-dimensional with 900 observations of random normal variables. We will separate the data into three classes, each with a different mean. 

```python
def get_simple_data():
    # assume 3 means
    D = 2 # so we can visualize it more easily
    s = 4 # separation so we can control how far apart the means are
    mu1 = np.array([0, 0])
    mu2 = np.array([s, s])
    mu3 = np.array([0, s])

    N = 900 # number of samples
    X = np.zeros((N, D))
    X[:300, :] = np.random.randn(300, D) + mu1
    X[300:600, :] = np.random.randn(300, D) + mu2
    X[600:, :] = np.random.randn(300, D) + mu3
    return X
```

Our distance metric is going to be squared distance between our vectors.

```python
def d(u, v):
    diff = u - v
    return diff.dot(diff)
```

The cost function we are trying to minimize is the squared distance between every point and every cluster, weighted by the responsibility of that point to that cluster. The inputs to this function are the observations, their responsibilities, and the means of each cluster.

```python
def cost(X, R, M):
    cost = 0
    for k in xrange(len(M)):
        # method 1
        # for n in xrange(len(X)):
        #     cost += R[n,k]*d(M[k], X[n])

        # method 2
        diff = X - M[k]
        sq_distances = (diff * diff).sum(axis=1)
        cost += (R[:,k] * sq_distances).sum()
    return cost
```

Our main function will execute the soft k-means procedure. 

```python
def plot_k_means(X, K, max_iter=20, beta=1.0, show_plots=True):
    N, D = X.shape
    M = np.zeros((K, D))
    exponents = np.empty((N, K))

    # initialize M to random
    for k in xrange(K):
        M[k] = X[np.random.choice(N)]

    costs = np.zeros(max_iter)
    for i in xrange(max_iter):
        # step 1: determine assignments / resposibilities
        for k in xrange(K):
            for n in xrange(N):
                exponents[n,k] = np.exp(-beta*d(M[k], X[n]))

        R = exponents / exponents.sum(axis=1, keepdims=True)

        # step 2: recalculate means
        for k in xrange(K):
            M[k] = R[:,k].dot(X) / R[:,k].sum()

        costs[i] = cost(X, R, M)
        if i > 0:
            if np.abs(costs[i] - costs[i-1]) < 10e-5:
                break

    if show_plots:
        plt.figure(figsize=(8,6))
        plt.plot(costs)
        plt.title("Costs")
        plt.show()

        random_colors = np.random.random((K, 3))
        colors = R.dot(random_colors)
        plt.figure(figsize=(8,6))
        plt.scatter(X[:,0], X[:,1], c=colors)
        plt.show()

    return M, R
```

```python
X = get_simple_data()

# what does it look like without clustering?
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
K = 3 # luckily, we already know this
plot_k_means(X, K)
```
